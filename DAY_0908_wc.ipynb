## 연도별 분석

```python
# 모듈 로딩
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import pandas as pd


year_song = list()
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36'}
for i in range(1970, 2022):
  
    url= f'https://www.genie.co.kr/chart/musicHistory?year='+str(i)+'&category=0'
    html = requests.get(url,headers=headers)
    soup = BeautifulSoup(html.text, 'html.parser')
    div = soup.find('div', {'id':'wrap-body'})
    div_input = div.find('input', {'id':'sAllSongID'})
    div_input_value = div_input['value']
    url_list = div_input_value.split(';')  # 노래의 주소 담기

    gasa = list()
    for k in url_list[:-1]:  # 노래의 가사 담기
        url1 = f'https://www.genie.co.kr/detail/songInfo?xgnm='+str(k)
        html1 = requests.get(url1,headers=headers)
        soup1 = BeautifulSoup(html1.text, 'html.parser')
        div1 = soup1.find('div',{'class':'tit-box'})
        div_p1 = div1.find('p').text
        
        gasa.append(div_p1)
        
    year_song.append(gasa)
```

```python
# 1. 명사에 해당하는 단어 추출, 수집
from konlpy.tag import Okt
from collections import Counter
okt = Okt()  # Open Korean Text 객체 생성

# okt함수를 통해 읽어 들인 내용의 형태소를 분석한다.
sentences_tag = []
sentences_tag = okt.pos(''.join(year_song[0]+year_song[1]+year_song[2]+year_song[3]+year_song[4]+year_song[5]+year_song[6]+year_song[7]+year_song[8]+year_song[9]))

noun_adj_list = []
# tag가 명사이거나 형용사인 단어들만 noun_addj_list에 넣어준다.
for word, tag in sentences_tag:
    if tag in ['Noun']:
        noun_adj_list.append(word)
        
# 가장 많이 나온 단어부터개를 저장한다.
counts = Counter(noun_adj_list)
tags = counts.most_common(100)
print(tags)
```

```python
# 모듈 로딩
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests

# 장르별 주소 추출
url = f'https://www.genie.co.kr/chart/genre?ditc=W&ymd=&genrecode=M0100'
html = requests.get(url,headers=headers)
soup = BeautifulSoup(html.text, 'html.parser')

div = soup.find('div',{'class':'this-type'})
div_a = div.find_all('a')

url_list = list()  # 장르 주소 담기
for i in div_a:
    url_list.append(i['href']) 

# 날짜 추출
dates = pd.date_range('2020-01','2022-08',freq='YS')

date = []  # 월요일 날짜 담기
for i in dates:
    date.append(str(i)[:10].replace('-', ''))

    
gg_song = list()    
for gern in url_list:  # 장르별
    gasa_ = [] 
    
    for week in date:
        url1 = f'https://www.genie.co.kr/chart/genre'+gern+'&ymd='+str(week)
        html1 = requests.get(url1,headers=headers)
        soup1 = BeautifulSoup(html1.text, 'html.parser')
        
        # 노래 주소 추출
        div1 = soup.find('div', {'class':'chart-date'})
        div1_input = div1.find('input', {'id':'sAllSongID'})['value']

        song_url = div1_input.split(';')
        song_url
        for su in song_url:
            url2 = f'https://www.genie.co.kr/detail/songInfo?xgnm='+su
            html2 = requests.get(url2,headers=headers)
            soup2 = BeautifulSoup(html2.text, 'html.parser')
            div2 = soup2.find('div',{'class':'tit-box'})
            div_p2 = div2.find('p').text
            
            gasa_.append(div_p2)
            
    gg_song.append(gasa_)
```
