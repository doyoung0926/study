```python
# 모듈 로딩
import pandas as pd
import numpy as np

# 합계 행 담기
data = list()
for i in range(2014, 2022):
    for j in range(1,13):
        df = pd.read_csv(str(i)+str(j)+'.csv', encoding = 'cp949')
        
        data.append(df[df['풍속'] == '합계'])
        
# 하나의 데이터 프레임 만들기        
DF = pd.concat(data)
```

```python
# 서쪽 방향 컬럼명 담기
w = list()
for col in DF.columns:
    if '서' in col:
        w.append(col)

# 서쪽 방향 합
DF['w_sum'] = DF[w].apply(np.sum, axis = 1)     
```

```python
# 날짜 컬럼 넣기
date = pd.date_range('2014-01', '2021-12', freq = 'MS')

month = list()
for i in date:
    month.append(i.to_period('m'))
    
DF['날짜'] = month
DF
```

```python
DF.to_csv('서울바람방향.csv')
```
