```python
# 모듈 로딩
import pandas as pd
```

```python
# 파일 로딩
df = pd.read_csv('구글.csv', header = 1)
df.info()
```

```python
df
```

```python
# 컬럼명 변경
df.columns=['날짜', '마스크', '공기청정기']
df
```

```python
###########################################3333
```

```python
# 파일 로딩
df1 = pd.read_excel('네이버.xlsx', header = 6)
df1.info()
```

```python
df1
```

```python
# 날짜 컬럼 하나 삭제
df1.drop('날짜.1', axis = 1, inplace = True)
```

```python
df1
```

```python
################################
```

```python
# 하나의 데이터프레임 합치기
DF = pd.concat([df, df1])
DF.info()
```

```python
# 날짜 컬럼 형 변환
DF['날짜'] = pd.to_datetime(DF['날짜'])

month = list()
for i in DF['날짜']:
    month.append(i.to_period('M'))
    
DF['날짜'] = month
```

```python
DF.info()
```

```python
DF.to_csv('검색어.csv')
```
