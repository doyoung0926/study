```python
# 모듈 로딩
import pandas as pd
```

```python
# 파일 로딩
df= pd.read_excel('석유소비점유율.xlsx')  # 석탄소비점유율.xlsx로 바꾸기
df.info()
```

```python
# 데이터 컬럼 결측치 개수 파악
df['데이터'].isnull().sum()
```

```python
# 데이터 컬럼 결측치 0으로 채우기
df['데이터'].fillna(0, inplace = True)
```

```python
# 상위 값으로 채우기
df.fillna(method = 'ffill', inplace = True)
```

```python
# 국가별 컬럼 값 확인
df['국가별'].unique()
```

```python
# 국가별 컬럼 칸 제거
country = list()
for i in df['국가별']:
    country.append(i.strip())
    
df['국가별'] = country

df['국가별'].unique()
```

```python
# 대한민국 중국만 추출
kor = df['국가별'] == '대한민국'
chi = df['국가별'] == '중국'

DF = df[kor | chi]
```

```python
# 인덱스 제설정
DF.index = range(0, DF.shape[0])
DF
```

```python
DF['항목'].unique()
```

```python
# 소비와 점유율로 따로 담기
a = list(DF['데이터'])[0::2] # 소비
b = list(DF['데이터'])[1::2] # 점유율
```

```python
# 원하는 형태로 만들기
DF_ = DF[DF['항목'] == '- 점유율 (%)']
DF_.drop(['항목', '데이터'], axis = 1, inplace = True)
```

```python
DF_['소비'] = a
DF_['점유율'] = b
```

```python
DF_.columns = ['날짜', '국가', '소비', '점유율']
```

```python
DF_.index = range(DF_.shape[0])
```

```python
date=pd.date_range('1995', '2020', freq = 'YS')

month = []
for i in date:
    month.append(i.to_period('Y'))
    month.append(i.to_period('Y'))
    
DF_['날짜'] = month
```

```python
DF_
```

```python
DF_.to_csv('석유소비점유율.csv')  # 석탄소비점유율.csv로 바꾸기
```
