```python
# 모듈 로딩
import pandas as pd
```

```python
# 파일 로딩
df = pd.read_csv('서울황사일수.csv', encoding = 'cp949')
df
```

```python
# 데이터 추출
data = []

for i in range(df.shape[0]):
    for j in range(1, df.shape[1]):
        data.append(df.iloc[i,j])

# 원하는 날짜 추출
date = pd.date_range('1958-01', '2021-12', freq='MS')

month = []
for k in date:
    month.append(k.to_period('m'))
month
```

```python
# 원하는 데이터프레임 만들기
month = pd.Series(month)
data = pd.Series(data)

DF = pd.DataFrame({'날짜': month,'서울황사일수': data})
DF.info()
```

```python
DF.to_csv('서울황사일수.csv')
```
