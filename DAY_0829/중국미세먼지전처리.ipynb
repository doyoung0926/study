```python
# 모듈 로딩
import pandas as pd
```

```python
# 파일 로딩
df = pd.read_csv('2014.txt', header = 12)
df = df.T
```

```python
# 일단 데이터 프레임 만들기
a = []
for i in range(0, 23, 2):
    tem = df[[0+i,1+i]]
    tem.columns = [0, 1]
    a.append(tem)

DF = pd.concat(a)
DF
```

```python
# 날짜 컬럼 넣기
date = pd.date_range('2014-01', '2014-12', freq='MS')

month = []
for i in date:
    month.append(i.to_period('M'))
    
DF.insert(0, '날짜', month)
DF.info()
```

```python
# 인덱스 및 컬럼명 설정
DF.index = range(0, DF.shape[0])
DF.columns = ['날짜', '매우 나쁨', '나쁨']
```

```python
# 파일명 담기
FILE = list()
for i in range(2014, 2022):
    FILE.append(str(i)+'.txt')
    
FILE
```

```python
value = list()
for file in FILE:
    df = pd.read_csv(file, header = 12)
    df = df.T
    
    a = []
    for i in range(0, 23, 2):
        tem = df[[0+i,1+i]]
        tem.columns = [0, 1]
        a.append(tem)

    DF = pd.concat(a)
    
    date = pd.date_range(str(file[:4])+'-01', str(file[:4])+'-12', freq='MS')

    month = []
    for i in date:
        month.append(i.to_period('M'))

    DF.insert(0, '날짜', month)
    
    DF.index = range(0, DF.shape[0])
    DF.columns = ['날짜', '매우 나쁨', '나쁨']
    
    value.append(DF)
```

```python
DF_ = pd.concat(value)
DF_
```

```python
import numpy as np 
sum = DF_.loc[:,['매우 나쁨', '나쁨']].apply(np.sum, axis=1)

DF_['bad_sum'] = sum
```
