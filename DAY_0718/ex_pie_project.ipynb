# 공공데이터 과제 #3
- 1. 대구시의 7개 구별 (중구, 동구, 서구, 남구, 북구, 수성구, 달서구) 남녀 비율 및 대구시 전체의 남녀 비율을 각각의 파이 차트로 나타내시오.
- subplots를 이용하여 4x2 형태의 총 8개의 subplot을 파이 차트로 구현
- gender.csv 파일 사용

```python
import csv
import matplotlib.pyplot as plt
import platform
   
f=open('gender.csv', encoding='euc_kr')
data=csv.reader(f)
city_list=['','중구','동구','서구','남구','북구','수성구','달서구']  # city_list 파일의 순서대로 입력!

male_sum=[]
female_sum=[]
for city in city_list:
    male_list=[] # 기존 리스트 데이터 초기화
    female_list=[] # 기존 리스트 데이터 초기화
    for row in data:
        if '대구광역시'+' '+city in row[0]:
            people=row
            for i in range(106, 207):
                male_list.append(int(people[i].replace(',','')))  # 천단위 콤마 제거
                female_list.append(int(people[i+103].replace(',','')))  # 천단위 콤마 제거
            break
    male_sum.append(sum(male_list))
    female_sum.append(sum(female_list))

if platform.system()=='Windows':
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='AppleGothic')

# 파이그래프 그리기    
fig, axes=plt.subplots(2, 4, figsize=(15, 10))
fig.suptitle('대구광역시 구별 남녀 인구 비율', fontsize=25)

for i in range(1,8):
    plt.subplot(2, 4, i)
    color=['cornflowerblue', 'tomato']
    plt.title('대구광역시'+' '+city_list[i])
    plt.pie([male_sum[i], female_sum[i]], labels=['남성', '여성'], autopct='%.1f%%', colors=color, startangle=90) # 파이 차트의 시작 각도 설정(90도)

plt.subplot(2, 4, 8)
plt.title('대구광역시 전체')
color=['cornflowerblue', 'tomato']
plt.pie([male_sum[0], female_sum[0]], labels=['남성', '여성'], autopct='%.1f%%', colors=color, startangle=90) # 파이 차트의 시작 각도 설정(90도)
plt.show()   # plt.show() 는 마지막에 한번만 하기!!
```
