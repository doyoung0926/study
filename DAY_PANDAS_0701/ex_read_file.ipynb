# 2. ‘iris.csv’에서 아래 결과를 도출하세요.

```python
# 모듈 로딩 ----------------------------------------------------------------------------
import pandas as pd

# 파일 관련 변수들 선언 ---------------------------------------------------------------
DIR_PATH='../Data/'
FILE_NAME=DIR_PATH+'iris.csv'

# CSV 파일 읽기 ===> DataFrame으로 저장 ----------------------------------------------
irisDF=pd.read_csv(FILE_NAME)

# 데이터 요약 정보 확인 -----------------------------------------------------------------------
irisDF.info()
```

## 1. Length 데이터만 추출해서 출력해주세요.

```python
# Length 데이터만 추출해서 출력
df=irisDF.drop(['SepalWidth','PetalWidth','Name'],axis=1)
df
```

## 2. SepalLength 컬럼에서 길이가 5.0보다 큰 컬럼의 SepalLength와 Names만 출력해주세요.

```python
# SepalLength 컬럼에서 길이가 5.0보다 큰 컬럼 찾기
SepalLengthCol=irisDF['SepalLength']
result=SepalLengthCol>5.0
df1=irisDF[result]
df1
```

```python
# SepalLength 컬럼에서 길이가 5.0보다 큰 컬럼의 SepalLength와 Names만 출력
df2=df1[['SepalLength','Name']]
df2
```

## 3. Iris-setosa, Iris-versicolor, Iris-virginica에서 각각 최대, 최소 컬럼을 출력해 주세요.

```python
# Iris-setosa에서 각각 최대, 최소 컬럼을 출력
NameCol=irisDF['Name']
setosa=NameCol=='Iris'+'-'+'setosa'
df3=irisDF[setosa].describe()
print(df3.loc['max'])
print(df3.loc['min'])
```

```python
# Iris-versicolor에서 각각 최대, 최소 컬럼을 출력
NameCol=irisDF['Name']
versicolor=NameCol=='Iris'+'-'+'versicolor'
df4=irisDF[versicolor].describe()
print(df4.loc['max'])
print(df4.loc['min'])
```

```python
# Iris-virginica에서 각각 최대, 최소 컬럼을 출력
NameCol=irisDF['Name']
virginica=NameCol=='Iris'+'-'+'virginica'
df5=irisDF[virginica].describe()
print(df5.loc['max'])
print(df5.loc['min'])
```
