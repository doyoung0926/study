## 모듈 로딩

```python
from scikeras.wrappers import KerasRegressor, KerasClassifier
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, InputLayer
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.utils import set_random_seed
from tensorflow.keras.backend import clear_session
import numpy as np
```

```python
## 동일 결과 출력 시드 설정
```

```python
clear_session()
np.random.seed(42)
set_random_seed(42)
```

## [1] 데이터 로딩 및 준비

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
x_data=data.data
y=data.target
```

```python
x_data.shape, y.shape
```

## [2] 모델 생성

```python
# 파라미터 => KerasClassifier 또는 KerassRegrassor의 parameter로 추가됨
def build_model(n_hidden, n_neurons, learning_rate, input_shape=(4,)):
    
    model = Sequential()
    
    model.add(InputLayer(input_shape=input_shape))
    
    for layer in range(n_hidden):
        model.add(Dense(n_neurons, activation='relu'))
        
    model.add(Dense(3, activation='softmax'))
    
    return model
```

```python
keras_clf = KerasClassifier(build_model,
                            n_hidden=1,
                            n_neurons=30,
                            learning_rate=0.001,
                            loss='sparse_categorical_crossentropy',
                            epochs=5)
```

```python
keras_clf.fit(x_data, y)
```

```python
mse_test = keras_clf.score(x_data, y)
```

```python
print(f'mse_test : {mse_test}')
```

```python
keras_clf.get_params
```

## [3] Hyperparameter Tunning

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
params = {
    'n_hidden': [1,3,5,10],
    'n_neurons': [10,30,50,70],
    'optimizer': ['adam','sgd'],
    'learning_rate': [0.0001,0.001,0.01,0.1],
}
gs = GridSearchCV(keras_clf, params,
                  refit=False, cv=3,
                  scoring='accuracy',
                  verbose=False)
```

```python
print('fitting...')
gs.fit(x_data, y)
print('END')
```

```python
print(gs.best_score_, gs.best_params_, sep='\n')
```
